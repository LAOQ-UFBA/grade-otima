```python
def resolve_model(
    courses,
    prereqs,
    course_names,
    even_semesters_courses,
    odd_semesters_courses,
    max_courses_per_semester,
    completed_courses,
    first_semester_courses
):
    
    semesters = list(range(1, 13))  

    remaining_courses = [c for c in courses if c not in completed_courses]
    prereqs = {c: [p for p in prereqs.get(c, []) if p in remaining_courses] for c in remaining_courses}

    model = pulp.LpProblem('GradeCurricular', pulp.LpMinimize)
    x = pulp.LpVariable.dicts('x', (remaining_courses, semesters), cat='Binary')

    # ===============================================================
    # RESTRIÇÕES
    # ===============================================================

    # 1. Cada disciplina deve ir para exatamente um semestre
    for c in remaining_courses:
        model += pulp.lpSum(x[c][s] for s in semesters) == 1

    if len(completed_courses) == 0:
        # 2. Fixar disciplinas obrigatórias do 1º semestre
        # (essas matérias são pré-definidas na matriz curricular)
        for c in first_semester_courses:
            model += x[c][1] == 1

        # 3. Garantir que apenas essas disciplinas estejam no 1º semestre
        # (impede o solver de colocar outras matérias junto delas)
        model += pulp.lpSum(x[c][1] for c in courses) == len(first_semester_courses)

    # 4. Máximo de matérias por semestre
    for s in semesters:
        model += pulp.lpSum(x[c][s] for c in remaining_courses) <= max_courses_per_semester

    # 5. Pré-requisitos
    for course, prereq_list in prereqs.items():
        for prereq in prereq_list:
            for s in semesters:
                model += x[course][s] <= pulp.lpSum(x[prereq][t] for t in semesters if t < s)

    # 6. Paridade
    for c in remaining_courses:
        if c in even_semesters_courses:
            allowed = [s for s in semesters if is_even(s)]
            model += pulp.lpSum(x[c][s] for s in allowed) == 1
        elif c in odd_semesters_courses:
            allowed = [s for s in semesters if not is_even(s)]
            model += pulp.lpSum(x[c][s] for s in allowed) == 1

    # Variáveis indicando uso real de cada semestre (se contém matéria da paridade)
    y_odd  = pulp.LpVariable.dicts('y_odd',  [s for s in semesters if s % 2 == 1], cat='Binary')
    y_even = pulp.LpVariable.dicts('y_even', [s for s in semesters if s % 2 == 0], cat='Binary')

    # Liga semestre ímpar ao uso real
    for s in y_odd:
        model += y_odd[s] >= pulp.lpSum(x[c][s] for c in remaining_courses if c in odd_semesters_courses) / max_courses_per_semester

    # Liga semestre par ao uso real
    for s in y_even:
        model += y_even[s] >= pulp.lpSum(x[c][s] for c in remaining_courses if c in even_semesters_courses) / max_courses_per_semester

    # Impede buracos na sequência de semestres ímpares
    odd_list = sorted([s for s in semesters if s % 2 == 1])
    for i in range(1, len(odd_list)):
        model += y_odd[odd_list[i]] <= y_odd[odd_list[i-1]]

    # Impede buracos na sequência de semestres pares
    even_list = sorted([s for s in semesters if s % 2 == 0])
    for i in range(1, len(even_list)):
        model += y_even[even_list[i]] <= y_even[even_list[i-1]]

    # ===============================================================
    # FUNÇÃO OBJETIVO
    # ===============================================================

    # Puxa cada matéria para o menor semestre possível
    model += pulp.lpSum(s * x[c][s] for c in remaining_courses for s in semesters)

    # ===============================================================
    # SOLUÇÃO
    # ===============================================================

    model.solve()

    rows = []

    for s in semesters:
        courses_in_semester = [c for c in remaining_courses if x[c][s].value() == 1]

        if courses_in_semester:
            tipo = "Ímpar" if s % 2 != 0 else "Par"

            for c in courses_in_semester:
                rows.append({
                    "Semestre": s,
                    "Tipo": tipo,
                    "Código": c,
                    "Disciplina": course_names.get(c, c)
                })

    return pd.DataFrame(rows)
```

```python
def resolve_model(
    courses,
    prereqs,
    course_names,
    even_semesters_courses,
    odd_semesters_courses,
    max_courses_per_semester,
    completed_courses,
    first_semester_courses
):

    semesters = list(range(1, 13))

    # ===============================================================
    # Filtra disciplinas ainda não concluídas
    # ===============================================================
    remaining_courses = [c for c in courses if c not in completed_courses]

    # Remove pré-requisitos já cumpridos
    prereqs = {
        c: [p for p in prereqs.get(c, []) if p in remaining_courses]
        for c in remaining_courses
    }

    # ===============================================================
    # MODELO
    # ===============================================================
    model = pulp.LpProblem('GradeCurricular', pulp.LpMinimize)
    x = pulp.LpVariable.dicts('x', (remaining_courses, semesters), cat='Binary')

    # ===============================================================
    # RESTRIÇÕES
    # ===============================================================

    # 1. Cada disciplina deve ir para exatamente um semestre
    for c in remaining_courses:
        model += pulp.lpSum(x[c][s] for s in semesters) == 1

    # 2. Regras do primeiro semestre (somente se não cursou nada ainda)
    if len(completed_courses) == 0:

        # Fixar disciplinas obrigatórias no 1º semestre
        for c in first_semester_courses:
            if c in remaining_courses:
                model += x[c][1] == 1

        # Nenhuma outra matéria pode entrar no semestre 1
        obrigatorias = [c for c in first_semester_courses if c in remaining_courses]
        model += pulp.lpSum(x[c][1] for c in remaining_courses) == len(obrigatorias)

    # 3. Máximo de matérias por semestre
    for s in semesters:
        model += pulp.lpSum(x[c][s] for c in remaining_courses) <= max_courses_per_semester

    # 4. Pré-requisitos
    for course, prereq_list in prereqs.items():
        for prereq in prereq_list:
            for s in semesters:
                model += x[course][s] <= pulp.lpSum(
                    x[prereq][t] for t in semesters if t < s
                )

    # 5. Paridade (semestres pares/ímpares)
    for c in remaining_courses:
        if c in even_semesters_courses:
            allowed = [s for s in semesters if s % 2 == 0]
            model += pulp.lpSum(x[c][s] for s in allowed) == 1

        elif c in odd_semesters_courses:
            allowed = [s for s in semesters if s % 2 == 1]
            model += pulp.lpSum(x[c][s] for s in allowed) == 1

    # ===============================================================
    # Variáveis auxiliares para evitar "buracos" por paridade
    # ===============================================================
    y_odd  = pulp.LpVariable.dicts('y_odd',  [s for s in semesters if s % 2 == 1], cat='Binary')
    y_even = pulp.LpVariable.dicts('y_even', [s for s in semesters if s % 2 == 0], cat='Binary')

    # liga y_odd
    for s in y_odd:
        model += y_odd[s] >= (
            pulp.lpSum(x[c][s] for c in remaining_courses if c in odd_semesters_courses)
            / max_courses_per_semester
        )

    # liga y_even
    for s in y_even:
        model += y_even[s] >= (
            pulp.lpSum(x[c][s] for c in remaining_courses if c in even_semesters_courses)
            / max_courses_per_semester
        )

    # impede buracos ímpares
    odd_list = sorted(y_odd.keys())
    for i in range(1, len(odd_list)):
        model += y_odd[odd_list[i]] <= y_odd[odd_list[i - 1]]

    # impede buracos pares
    even_list = sorted(y_even.keys())
    for i in range(1, len(even_list)):
        model += y_even[even_list[i]] <= y_even[even_list[i - 1]]

    # ===============================================================
    # FUNÇÃO OBJETIVO
    # ===============================================================
    # Minimiza semestres atribuídos, puxando tudo para o mais cedo possível
    model += pulp.lpSum(s * x[c][s] for c in remaining_courses for s in semesters)

    # ===============================================================
    # SOLUÇÃO
    # ===============================================================
    model.solve()

    # ===============================================================
    # MONTA DATAFRAME RESULTADO
    # ===============================================================
    rows = []

    for s in semesters:
        courses_in_semester = [c for c in remaining_courses if x[c][s].value() == 1]

        if courses_in_semester:
            tipo = "Ímpar" if s % 2 != 0 else "Par"

            for c in courses_in_semester:
                rows.append({
                    "Semestre": s,
                    "Tipo": tipo,
                    "Código": c,
                    "Disciplina": course_names.get(c, c)
                })

    return pd.DataFrame(rows)
```

```python
def resolve_model(
    courses,
    prereqs,
    course_names,
    even_semesters_courses,
    odd_semesters_courses,
    max_courses_per_semester,
    completed_courses,
    first_semester_courses
):

    semesters = list(range(1, 13))

    # ===============================================================
    # Filtra disciplinas que ainda faltam
    # ===============================================================
    remaining_courses = [c for c in courses if c not in completed_courses]

    # Ajusta pré-requisitos para considerar somente o que ainda existe
    prereqs = {
        c: [p for p in prereqs.get(c, []) if p in remaining_courses]
        for c in remaining_courses
    }

    # ===============================================================
    # MODELO DE OTIMIZAÇÃO
    # ===============================================================
    model = pulp.LpProblem('GradeCurricular', pulp.LpMinimize)
    x = pulp.LpVariable.dicts('x', (remaining_courses, semesters), cat='Binary')

    # ===============================================================
    # RESTRIÇÕES BASE
    # ===============================================================

    # 1. Cada disciplina é alocada em exatamente um semestre
    for c in remaining_courses:
        model += pulp.lpSum(x[c][s] for s in semesters) == 1

    # 2. Limite máximo de matérias por semestre
    for s in semesters:
        model += pulp.lpSum(x[c][s] for c in remaining_courses) <= max_courses_per_semester

    # 3. Restrições de pré-requisito
    for course, prereq_list in prereqs.items():
        for prereq in prereq_list:
            for s in semesters:
                model += x[course][s] <= pulp.lpSum(
                    x[prereq][t] for t in semesters if t < s
                )

    # 4. Restrições de paridade (pares / ímpares)
    for c in remaining_courses:

        # disciplina obrigatoriamente em semestre PAR
        if c in even_semesters_courses:
            allowed = [s for s in semesters if s % 2 == 0]
            model += pulp.lpSum(x[c][s] for s in allowed) == 1

        # disciplina obrigatoriamente em semestre ÍMPAR
        elif c in odd_semesters_courses:
            allowed = [s for s in semesters if s % 2 == 1]
            model += pulp.lpSum(x[c][s] for s in allowed) == 1

    # ===============================================================
    # REGRAS DO PRIMEIRO SEMESTRE
    # ===============================================================
    if len(completed_courses) == 0:

        # Fixar obrigatórias no semestre 1
        for c in first_semester_courses:
            if c in remaining_courses:
                model += x[c][1] == 1

        # Nenhuma outra matéria pode entrar no semestre 1
        obrigatorias = [c for c in first_semester_courses if c in remaining_courses]
        model += pulp.lpSum(x[c][1] for c in remaining_courses) == len(obrigatorias)

    # ===============================================================
    # VARIÁVEL PARA MINIMIZAR O ÚLTIMO SEMESTRE OCUPADO
    # ===============================================================
    last_semester = pulp.LpVariable(
        'ultimo_semestre',
        lowBound=1,
        upBound=max(semesters),
        cat='Integer'
    )

    # last_semester >= semestre real de cada disciplina
    for c in remaining_courses:
        for s in semesters:
            model += last_semester >= s * x[c][s]

    # ===============================================================
    # FUNÇÃO OBJETIVO
    # Minimizar o último semestre ocupado
    # ===============================================================
    model += last_semester

    # ===============================================================
    # SOLUÇÃO
    # ===============================================================
    model.solve()

    # ===============================================================
    # MONTA DATAFRAME
    # ===============================================================
    rows = []

    for s in semesters:
        cursos = [c for c in remaining_courses if x[c][s].value() == 1]

        if cursos:
            tipo = "Ímpar" if s % 2 != 0 else "Par"

            for c in cursos:
                rows.append({
                    "Semestre": s,
                    "Tipo": tipo,
                    "Código": c,
                    "Disciplina": course_names.get(c, c)
                })

    df = pd.DataFrame(rows).sort_values("Semestre")

    return df, last_semester.value()
```
